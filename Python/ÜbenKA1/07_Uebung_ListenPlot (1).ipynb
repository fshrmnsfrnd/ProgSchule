1. Erstellen Sie eine Liste x mit Zahlen von -10 bis 10 in einem Abstand von 0.1

```python
i: float = -10
x: list = []
while(i <= 10):
    x.append(i)
    i += 0.1
```

### Erstellen Sie eine Liste y mit den Funktionswerten der Funktion:

    y = x² + 3x + 3

    Definitionsbereich ist die Liste aus Aufgabe 1
    

```python
i: float
y1: list = []
for i in x:
    e: float = pow(i, 2) + 3*i + 3
    y1.append(e)
```

### Lassen Sie den Graphen der Funktion anzeigen. Verwenden Sie das Modul mathplotlib mit der Klasse pyplot. Verwenden Sie im Programm das Kürzel plt für pyplot


Hier finden Sie ein Tutorial für matplotlib:<br>

und hier finden Sie die komplette Dokumentation:<br>
https://matplotlib.org/ 


```python
import matplotlib.pyplot as plt

plt.plot(x, y1)
```

### Erzeugen Sie einen Graphen für die Funktion:

> y = 2x +1



```python
i: float
y2: list = []
for i in x:
    e: float = 2*i + 2
    y2.append(e)

plt.plot(x, y2, label="dsafj")
plt.legend()
```

### Erzeugen Sie einen Graphen für die Funktion:

> y = e<sup>x </sup>



```python
import math

i: float
y3: list = []
for i in x:
    e: float = pow(math.e, i)
    y3.append(e)

plt.plot(x, y3)
```

### Erzeugen Sie einen Graphen für die Funktion:

> y = log<sub>2</sub> (x + 11)

Ändern Sie die Linienfarbe und beschriften Sie die Achsen. Zeichnen Sie die Funktion in einem Raster (Grid)



```python
i: float
y4: list = []
for i in x:
    e: float = math.log2(i+11)
    y4.append(e)
plt.plot(x, y4)
plt.grid()
```

### Zeichnen Sie alle 4 Graphen in einem 2* 2 Raster (plt.subplot)

```python
plt.subplot(2,2,1)
plt.grid()
plt.plot(x, y1)

plt.subplot(2,2,2)
plt.grid()
plt.plot(x, y2)

plt.subplot(2,2,3)
plt.grid()
plt.plot(x, y3)

plt.subplot(2,2,4)
plt.grid()
plt.plot(x, y4)
```
