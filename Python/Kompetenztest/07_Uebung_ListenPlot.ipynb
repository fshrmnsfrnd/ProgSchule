1. Erstellen Sie eine Liste x mit Zahlen von -10 bis 10 in einem Abstand von 0.1

```python
x = []

counter = -10
while counter <= 10:
    x.append(counter)
    counter += 0.1
```

### Erstellen Sie eine Liste y mit den Funktionswerten der Funktion:

    y = x² + 3x + 3

    Definitionsbereich ist die Liste aus Aufgabe 1
    

```python
y = []
for i in x:
    y.append(i**2 + 3*i + 3)
```

### Lassen Sie den Graphen der Funktion anzeigen. Verwenden Sie das Modul mathplotlib mit der Klasse pyplot. Verwenden Sie im Programm das Kürzel plt für pyplot

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
plt.show()
```

### Erzeugen Sie einen Graphen für die Funktion:

> y = 2x +1



```python
x = []
i = 0
while i <= 10:
    x.append(i)
    i += 1

y = []
for i in x:
    y.append(2*i + 1)

plt.plot(x, y)
plt.show()
```

### Erzeugen Sie einen Graphen für die Funktion:

> y = e<sup>x </sup>



```python
import math
x = []
i = 0
while i <= 10:
    x.append(i)
    i += 0.1

y = []
for i in x:
    y.append(math.e**i)

plt.plot(x, y)
plt.show()
```

### Erzeugen Sie einen Graphen für die Funktion:

> y = log<sub>2</sub> (x + 11)

Ändern Sie die Linienfarbe und beschriften Sie die Achsen. Zeichnen Sie die Funktion in einem Raster (Grid)



```python
x = []
i = 0
while i <= 1000:
    x.append(i)
    i += 1

y = []
for i in x:
    y.append(math.log(i + 11, 2))

plt.plot(x, y, color='red')
plt.grid(True, color='grey', linestyle='-', linewidth=1)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
